### Funció GELU

La funció GELU (Gaussian Error Linear Unit) es defineix com:

$$ GELU(x) = x \cdot \Phi(x) $$

on $\Phi(x)$ és la funció de distribució acumulativa de la distribució normal estàndard, que es pot expressar utilitzant la funció d'error (erf) de la següent manera:

$$ \Phi(x) = \frac{1}{2} \left(1 + \text{erf}\left(\frac{x}{\sqrt{2}}\right)\right) $$

Per tant, l'expressió completa de GELU utilitzant erf és:

$$ GELU(x) = 0.5 \cdot x \cdot \left(1 + \text{erf}\left(\frac{x}{\sqrt{2}}\right)\right) $$


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gelu(x):
    return 0.5 * x * (1 + np.tanh(np.sqrt(2 / np.pi) * (x + 0.044715 * x**3)))
```

```python
x = np.linspace(-5, 5, 100)

y = gelu(x)

plt.figure(figsize=(10, 5))
plt.plot(x, y, label="GELU approximation with tanh")
plt.title("GELU Activation Function")
plt.xlabel("x")
plt.ylabel("GELU(x)")
plt.grid(True)
plt.legend()
plt.show()
```
